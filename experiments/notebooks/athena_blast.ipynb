```python
from fvhoe.solver import EulerSolver
import matplotlib.pyplot as plt
import numpy as np
```

```python
solver = EulerSolver()
snapshot_dir = "/scratch/gpfs/jp7427/fvhoe/snapshots/"
solver.rkorder(1, snapshot_dir=snapshot_dir + "athena_blast_NAD=0.01")
```

```python
fig, ax = plt.subplots()

solver.plot_2d_slice(ax, param="rho", z=0.5, vmin=0.08, vmax=0.65, cmap="gnuplot2_r")
```

```python
rho = solver.snapshots[-1]["w"].rho[:, :, 0]

print(np.mean(np.abs(rho - np.flipud(rho))))
print(np.mean(np.abs(rho - np.fliplr(rho))))

plt.imshow(np.rot90(np.abs(rho - np.flipud(rho))))
plt.colorbar()
```
