```python
from fvhoe.boundary_conditions import BoundaryCondition
from fvhoe.initial_conditions import slotted_disk
from fvhoe.solver import EulerSolver
import matplotlib.pyplot as plt
import numpy as np
```

```python
def mx(x, y, z, t):
    yc = y - 0.5
    return -yc


def my(x, y, z, t):
    xc = x - 0.5
    return xc


def E(x, y, z, t):
    xc = x - 0.5
    yc = y - 0.5
    return 1 / (5 / 3 - 1) + 0.5 * (xc**2 + yc**2)


solver = EulerSolver(
    w0=slotted_disk,
    nx=64,
    ny=64,
    px=3,
    py=3,
    bc=BoundaryCondition(
        x="dirichlet",
        y="dirichlet",
        x_value={"rho": 1, "E": E, "mx": mx, "my": my, "mz": 0},
        y_value={"rho": 1, "E": E, "mx": mx, "my": my, "mz": 0},
    ),
    riemann_solver="advection_upwind",
    fixed_primitive_variables=["vx", "vy", "P"],
    a_posteriori_slope_limiting=True,
    slope_limiter="minmod",
    cupy=False,
)
```

```python
solver.run(2 * np.pi, downbeats=[0.5 * np.pi, np.pi, 1.5 * np.pi])
```

```python
fig, ax = plt.subplots(2, 2, sharex=True, sharey=True)

for var, label, idx in zip(
    ["rho", "P", "vx", "vy"],
    [r"$\overline{\rho}$", r"$\overline{P}$", r"$\overline{v}_x$", r"$\overline{v}_y$"],
    [(0, 0), (0, 1), (1, 0), (1, 1)],
):
    im = solver.plot_2d_slice(ax[idx], t=2 * np.pi, param=var, cmap="gnuplot", z=0.5)
    fig.colorbar(im, ax=ax[idx], label=label)

ax[0, 0].set_ylabel("$y$")
ax[1, 0].set_ylabel("$y$")
ax[1, 0].set_xlabel("$x$")
ax[1, 1].set_xlabel("$x$")
```

```python
for t in [0, 1]:
    m_tot = np.sum(solver.snapshots[0]["w"].rho[:, 0, 0]) * (
        solver.h[0] * solver.h[1] * solver.h[2]
    )
    print(f"Total mass at time {t}: {m_tot}")
```
