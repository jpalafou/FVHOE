```python
from fvhoe.initial_conditions import square
from fvhoe.solver import EulerSolver
```

1D square

```python
solver = EulerSolver(
    w0=square(),
    nx=128,
    px=3,
    riemann_solver="llf",
)
solver.rkorder(1.0)
solver.plot_fields(y=0.5, z=0.5)
```

2D square

```python
N = 64
p = 3

solver = EulerSolver(
    w0=square(dims="xy", vx=1, vy=1),
    nx=N,
    ny=N,
    px=p,
    py=p,
    riemann_solver="llf",
    cupy=False,
)
solver.rkorder(1.0)
solver.plot_fields(z=0.5)
```
