```python
from fvhoe.initial_conditions import square
from fvhoe.solver import EulerSolver
```

1D square

```python
solver = EulerSolver(
    w0=square(),
    nx=128,
    px=3,
    riemann_solver="llf",
)
solver.rkorder(1.0)
solver.plot_fields(y=0.5, z=0.5)
```

2D square

```python
solver = EulerSolver(
    w0=square(dims="xy"),
    nx=32,
    ny=32,
    px=3,
    py=3,
    riemann_solver="llf",
)
solver.rkorder(1.0)
solver.plot_fields(z=0.5)
```
