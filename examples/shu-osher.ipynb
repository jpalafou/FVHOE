```python
from fvhoe.boundary_conditions import BoundaryCondition
from fvhoe.config import conservative_names
from fvhoe.initial_conditions import shu_osher_1d
from fvhoe.solver import EulerSolver
import matplotlib.pyplot as plt
import numpy as np
```

```python
# set up numerical solution
N = 400
p = 3

solver = EulerSolver(
    w0=shu_osher_1d(),
    bc=BoundaryCondition(x="ic"),
    gamma=1.4,
    x=(0, 10),
    nx=N,
    px=p,
    riemann_solver="hllc",
    a_posteriori_slope_limiting=p > 0,
)

# run simulation
solver.rkorder(1.8)
```

```python
fig, ax = plt.subplots()
solver.plot_1d_slice(ax, param="rho", y=0.5, z=0.5)
ax.set_xlim(5, 7.5)
ax.set_xlabel(r"$x$")
ax.set_ylabel(r"$\rho$")
```
