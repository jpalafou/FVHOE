```python
from fvhoe.initial_conditions import kelvin_helmholtz_2d
from fvhoe.solver import EulerSolver
import matplotlib.pyplot as plt
```

```python
# set up numerical solution
N = 64
p = 3
snapshot_dir = f"../snapshots/kelvin-helmholtz-{N=}-{p=}"

solver = EulerSolver(
    w0=kelvin_helmholtz_2d,
    gamma=1.4,
    nx=N,
    ny=N,
    px=p,
    py=p,
    riemann_solver="hllc",
    a_posteriori_slope_limiting=p > 0,
    cupy=True,
)

# run simulation
solver.rkorder(0.8, snapshot_dir=snapshot_dir)
```

```python
fig, ax = plt.subplots()
solver.plot_2d_slice(ax, param="rho", z=0.5)
```
