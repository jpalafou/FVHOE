```python
from fvhoe.initial_conditions import kelvin_helmholtz_2d
from fvhoe.scripting import EulerSolver_wrapper
from fvhoe.solver import EulerSolver
import matplotlib.pyplot as plt
```

```python
# set up numerical solution
N = 128
p = 3

solver = EulerSolver_wrapper(
    project_pref="kelvin-helmholtz",
    snapshot_parent_dir="../../out",
    ic=kelvin_helmholtz_2d(),
    gamma=1.4,
    nx=N,
    ny=N,
    px=p,
    py=p,
    riemann_solver="hllc",
    a_posteriori_slope_limiting=p > 0,
    cupy=True,
    T=0.8,
)
```

```python
fig, ax = plt.subplots()
solver.plot_2d_slice(ax, param="rho", z=0.5)
```
