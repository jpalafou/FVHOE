```python
from fvhoe.boundary_conditions import BoundaryCondition
from fvhoe.initial_conditions import double_shock_1d
from fvhoe.solver import EulerSolver
import matplotlib.pyplot as plt
```

```python
# set up numerical solution
N = 240
p = 3

solver = EulerSolver(
    w0=double_shock_1d,
    bc=BoundaryCondition(x="reflective"),
    gamma=1.4,
    nx=N,
    px=p,
    riemann_solver="hllc",
    a_posteriori_slope_limiting=p > 0,
    NAD=1e-5,
    density_floor=True,
    pressure_floor=True,
)

# run simulation
solver.rkorder(0.038)
```

```python
fig, ax = plt.subplots()
solver.plot_1d_slice(ax, param="rho", y=0.5, z=0.5)
ax.set_xlabel(r"$x$")
ax.set_ylabel(r"$\rho$")
```
